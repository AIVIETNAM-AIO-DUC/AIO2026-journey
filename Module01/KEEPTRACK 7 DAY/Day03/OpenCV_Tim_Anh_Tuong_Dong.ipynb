Bài toán tìm kiếm hình ảnh tương đồng:

- Đồng bộ các ảnh về cùng không gian màu và kích thước

- Chuyển đổi ma trận điểm ảnh thành vector toán học (flatten)

- Đo mức độ sai khác của các vector, sai càng ít thì độ tương đồng càng cao

```python
import cv2
import numpy as np

def process_image(image_path, size = (64,64)):
  #đọc ảnh đầu vào dưới dạng ảnh xám
  img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

  if img is None:
    raise ValueError("Không thể đọc ảnh")

  #thay đổi kích thước ma trận ảnh về chuẩn 64*64
  img_resize = cv2.resize(img, size)

  #hiển thị kích thước ma trận ảnh thực tế sai khi sử lý không gian
  print(f"{image_path}Kích thước ảnh sau khi xử lý: {img_resize.shape}")


  #flatten 2d thành 1d: 64*64 -> 4096
  vector = img_resize.flatten()

  # nomalize dải giá trị điểm ảnh từ [0, 255] -> [0, 1]
  vector_normalized = vector / 255.0

  return vector_normalized
```

Các hàm đo lường độ tương đồng vector: L1, L2 distance, cosine similarity

```python
def l1_distance(v1, v2):
  #tính tổng trị tuyệt đối sai lệch hai vector
  return np.sum(np.abs(v1-v2))

def l2_distance(v1, v2):
  #dùng hàm tính chuẩn linalg.norm để đo euclid
  return np.linalg.norm(v1-v2)

def cosine_similarity(v1, v2):
  #tính tích vô hướng chia tích độ dài hai vector
  dot_product = np.dot(v1, v2)
  norm_v1 = np.linalg.norm(v1)
  norm_v2 = np.linalg.norm(v2)
  return dot_product / (norm_v1 * norm_v2)
```

Đối chiếu ảnh truy vấn với tập dữ liệu

```python
import os

#định nghĩa các đường dẫn tệp tin và thư mục chứ dữ liệu
query_path = "query.jpg"
dataset_dir = "images_folder/"

#tạo vector đặc trưng cho ảnh truy vấn
query_vector = process_image(query_path)

min_distance = float("inf")
best_math = None

print("\nbắt đầu quét trên tập dữ liêu...")

#duyệt lần lượt các tệp tin trong thư mục
for filename in os.listdir(dataset_dir):
  if filename.endswith((".jpg", ".png", ".jpeg")):
    db_img_path = os.path.join(dataset_dir, filename)
    try:
      #chuyển data thành vector
      db_vector = process_image(db_img_path)

      #áp dụng hàm khoảng cách l2 để đo
      distance = l2_distance(query_vector, db_vector)
      print(f"Khoảng cách giữa {query_path} và {db_img_path}: {distance}")

      #khi thấy kết quả tốt hơn thì cập nhật
      if distance < min_distance:
        min_distance = distance
        best_match = db_img_path
    except ValueError as e:
      print(e)


print("\n --- KẾT LUẬN HỆ THÔNG --- ")
if best_match:
  print (f"Tep tin co độ tưong đồng cao nhắt: {best_match}")
  print (f"Khoang cach sai lach toi thieu ghi nhận: {min_distance: .4f}")
else:
  print ("Hệ thong khong tim thay du lieu hinh anh hop le trong thu muc.")
```
