```python
# dòng lệnh để viết ra thành 1 file .py
%%writefile main.py

import streamlit as st


# Tiêu đề app
st.title("Tính hàm lũy thừa")


# Giới thiệu
st.write("Nhập x và số mũ n để tính lũy thừa x^n")


# Nhập dữ liệu
x = st.number_input(
    "Nhập x", value=2
)

n = st.number_input(
    "Nhập n", value=3
)

# bấm nút và chạy các lệnh từ trên xuống
if st.button("Tính kết quả"):
    result = x ** n
    st.write("Kết quả:")
    st.success(
        f"{x}^{n} = {result}"
    )
```
