```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

#Lấy danh sách link từ page

```python
def get_link_dantri(url):
    #gửi request đến trang web và lấy object HTML
    html_text = requests.get(url, headers={'User-Agent': 'test'})
    #lấy nội dung HTML từ object HTML
    soup = BeautifulSoup(html_text.text, 'html.parser')
    urls = []
    #tìm tất cả các thẻ div có class là article-title
    div = soup.find(class_="article list")
    #lấy thẻ a trong các div và append link vào list urls
    for article in div.find_all(class_="article-item"):
        #lấy thẻ div có class là article-thumb
        div_tag = article.find(class_="article-thumb")
        #lấy thẻ a trong div_tag và lấy link từ thuộc tính href
        a_tag = div_tag.find('a', href=True)
        link = a_tag.attrs['href']
        #nối link với url gốc để tạo thành link đầy đủ
        url_input = urljoin(url, link)
        urls.append(url_input)
    return urls
```

```python
get_link_dantri("https://dantri.com.vn/kinh-doanh/trang-2.htm")
```

```python
def get_content_dantri(url):
    #gửi request đến trang web và lấy object HTML
    html_text = requests.get(url, headers={'User-Agent': 'test'})
    #lấy nội dung HTML từ object HTML
    soup = BeautifulSoup(html_text.text, 'html.parser')
    #tìm thẻ div có id là desktop-in-article
    div_tag = soup.find(id="desktop-in-article")
    article = []
    #lấy tất cả các thẻ p trong div_tag và append nội dung vào list article
    for p_tag in div_tag.find_all(lambda tag: tag.name == 'p'):
        article.extend(p_tag.stripped_strings)
    return " ".join(article)
```

```python
url = "https://dantri.com.vn/kinh-doanh/chinh-phu-yeu-cau-nghien-cuu-de-xuat-ngan-hang-nha-nuoc-bao-lanh-trai-phieu-20230721141159416.htm"
get_content_dantri(url)

#có thể sử dụng vòng lặp để lấy nội dung của tất cả các link trong list urls
```
